```python
from pypdf import PdfReader
from langchain_core.documents import Document
```

```python
reader = PdfReader("../data/documents/Reglement-interieur-ESTIN.pdf")
```

```python
print(len(reader.pages))
```

```python
document = []
for page_num in range(1, len(reader.pages)-1):
    page = reader.pages[page_num]
    text = page.extract_text()

    doc = Document(
        page_content = text,
        metadata = {
            'page': page_num + 1,
            'source': f'Reglement-interieur-ESTIN.pdf'
        }
    )
    document.append(doc)
print(len(document))
```

```python
full_text = ""
    
for doc in document:
    page_num = doc.metadata.get("page", "?")
    full_text += f"\n\n[PAGE {page_num}]\n\n"
    full_text += doc.page_content
```

```python
print(full_text)
```
